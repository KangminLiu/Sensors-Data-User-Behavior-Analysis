```python
n_clickSubmit = 0
n_formSubmit = 0
n_errorSubmit = 0
n_others = 0
for i in range(len(df_sensors)):
    if df_sensors["event"][i] == "clickSubmit":
        n_clickSubmit += 1
    elif df_sensors["event"][i] == "formSubmit":
        n_formSubmit += 1
    elif df_sensors["event"][i] == "errorSubmit":
        n_errorSubmit += 1
    else:
        n_others += 1
    
print("number of clickSubmit: %d" % n_clickSubmit)
print("number of formSubmit: %d" % n_formSubmit)
print("number of errorSubmit: %d" % n_errorSubmit)
print("number of others: %d" % n_others)
print("number of event: %d" % (n_clickSubmit+n_formSubmit+n_errorSubmit+n_others))
```

```python
# print out the columns which have less than 20% NAs

j = 0
for item in df_sensors.columns:
    not_null_ratio = df_sensors[item].count() / float(len(df_sensors))
    if (not_null_ratio) >= 0.8:
        j += 1
        print("%d - %s %f" % (j, item, float(not_null_ratio)))
```

```python
# convert the Unix time to the recongnizable time format

from datetime import datetime

time_formatted = []

for i in range(len(df_sensors)):
    ds = int(df_sensors["time"][i])
    ds /= 1000
    ds_new = datetime.utcfromtimestamp(ds).strftime('%Y-%m-%d %H:%M:%S')
    time_formatted.append(ds_new)

df_sensors["time_formatted"] = pd.Series(time_formatted)
```
